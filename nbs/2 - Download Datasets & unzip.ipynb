```python
import pandas as pd
```

```python
sms_spam_input_path = SMS_SPAM_DIR/ "SMSSpamCollection"
sms_df = pd.read_csv(sms_spam_input_path, sep = '\t', header = None)
sms_df.columns =['label', 'text']
sms_df.tail()
```

```python
yt_df = []

for path in YOUTUBE_SPAM_DIR.glob('*.csv'):
    df = pd.read_csv(path)
    df.rename(columns={"CLASS":"RAW LABEL", "CONTENT":"TEXT"}, inplace = True)
    df['LABEL'] = df['RAW LABEL'].apply(lambda x: 'spam' if str(x) == "1" else 'ham' )
    yt_df.append(df)
yt_df  = pd.concat(yt_df)
yt_df.tail()
```
